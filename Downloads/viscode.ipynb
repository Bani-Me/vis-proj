```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd 
```

```python
x = pd.read_excel('C:\Task_pandas\cleaned_employee_sample_data.xlsx')
DFx = pd.DataFrame(x)
```

```python
numB = 0
for i in x['Ethnicity']:
    if i == "Black":
        numB += 1

numA = 0
for i in x['Ethnicity']:
    if i == "Asian":
        numA += 1

numC = 0
for i in x['Ethnicity']:
    if i == "Caucasian":
        numC += 1

numL = 0
for i in x['Ethnicity']:
    if i == "Latino":
        numL += 1
```

```python
y = np.array([numL,numA,numB,numC])
myLabels = ["Latino", "Asian","Black","Caucasian"]
myExplode = [0, 0.2, 0, 0]

plt.pie(y, labels = myLabels, explode = myExplode)
plt.legend()
plt.show()
```

```python
Sal = DFx.groupby("Ethnicity")
SalRel = Sal["Annual Salary"].mean()


j = np.array(["Asian", "Black", "Caucasian", "Latino"])
plt.title("Average Salary/Ethnicity")
plt.bar(j, SalRel, color = ["orange", "green", "red", "blue"] )
plt.show()
```

```python
Con = DFx.groupby("Country")
ConRel = Con["Annual Salary"].mean()


b = np.array(["United States", "China", "Brazil"])
plt.title("Average Salary/Country")
plt.bar(b, ConRel, color = ["orange", "green", "red", "blue"] )
plt.show()
```
